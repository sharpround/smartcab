```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# 100 trials
success = pd.read_pickle('success_02.pickle')
success.hist(bins=range(0, 101, 10))
plt.xlabel("trial number")
plt.ylabel("# of successes")
plt.title("success frequency, 100 trials, bin width = 10")

plt.savefig('success_1000.png', dpi=240)
```

```python
conv = pd.read_pickle('convergence_02.pickle')
plt.plot(conv)
plt.hold(True)
# plt.plot([0.0, 100.0], [0.7, 0.1])

plt.savefig('convergence_100.png', dpi=240)
```
